```python
import pandas as pd
import matplotlib.pyplot as plt

# MAGIC COMMAND: Ensures Matplotlib plots are displayed inline in the notebook
%matplotlib inline
```

```python
# --- 1. Load the data ---
file_path = "D:\AI For Product Management\Machine Learning Foundation For Product Managers\Prjects\CSV files\Amazon Big Billion Sale 2025 -Oct Mobile Phones.csv"
try:
    df = pd.read_csv(file_path)
    print(f"--- Data Loading Successful: {len(df)} rows loaded ---\n")
except FileNotFoundError:
    # In a notebook, a clear error message is better than exiting
    print(f"!!! Error: File not found at {file_path}. Please ensure the file is uploaded or in the correct path.")
    # Stop execution if data is not loaded
    raise
```

```python
# --- 2. Data Cleaning and Preparation ---

# Function to safely convert price (which might be stored as string/object)
def clean_price(price):
    """Removes commas and converts to float, returns NaN if cleaning fails."""
    try:
        if isinstance(price, str):
            # Remove commas and convert to float
            return float(price.replace(',', ''))
        return float(price)
    except:
        return None

# Apply cleaning to Price and Review_Count columns
df['Price_Cleaned'] = df['Price'].apply(clean_price)
```

```python
# The Review_Count column should also be converted to numeric, handling potential string issues
# We coerce errors to NaN
df['Review_Count_Cleaned'] = pd.to_numeric(df['Review_Count'], errors='coerce')

# Drop rows where essential data is missing for analysis
df_cleaned = df.dropna(subset=['Price_Cleaned', 'Rating', 'Review_Count_Cleaned'])
print(f"--- Data Cleaning Complete: {len(df_cleaned)} rows used for analysis ---\n")
```

```python
# Table 1: Overall Descriptive Statistics
print("="*60)
print("TABLE 1: OVERALL SALES DESCRIPTIVE STATISTICS")
print("="*60)
desc_stats = df_cleaned[['Price_Cleaned', 'Rating', 'Review_Count_Cleaned']].describe()
# Rename index and columns for better presentation
desc_stats.index = ['Count', 'Mean', 'Std Dev', 'Min', '25th Perc.', 'Median', '75th Perc.', 'Max']
desc_stats.columns = ['Price (INR)', 'Rating', 'Review Count']
# In Jupyter, displaying the DataFrame object often looks better than printing to_string()
print(desc_stats.round(2))
print("\n")
```

```python
# Table 2: Top 10 Products by Review Count
print("="*60)
print("TABLE 2: TOP 10 PRODUCTS BY CUSTOMER REVIEWS")
print("="*60)
top_reviewed = df_cleaned.sort_values(by='Review_Count_Cleaned', ascending=False).head(10)

# Select and rename columns for the output table
top_reviewed_table = top_reviewed[['Product_Name', 'Price_Cleaned', 'Rating', 'Review_Count_Cleaned']].copy()
top_reviewed_table.columns = ['Product Name', 'Price (INR)', 'Rating', 'Review Count']

# Limit the product name length for better table printing
top_reviewed_table['Product Name'] = top_reviewed_table['Product Name'].apply(lambda x: x[:70] + '...' if len(x) > 70 else x)

# Display the DataFrame
print(top_reviewed_table)
print("\n")
```

```python
# --- 4. Generate Charts ---

# Chart 1: Top 10 Products by Review Count (Bar Chart)
print("="*60)
print("CHART 1: TOP 10 PRODUCTS BY REVIEW COUNT")
print("="*60)
plt.figure(figsize=(12, 6))
top_10_reviews = top_reviewed_table.head(10)
# Reverse the order for a horizontal bar chart
plt.barh(top_10_reviews['Product Name'][::-1], top_10_reviews['Review Count'][::-1], color='#3498db') # Blue color
plt.xlabel('Total Review Count', fontsize=12)
plt.title('Top 10 Mobile Products by Customer Review Count', fontsize=14, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show() # This displays the chart inline

# Chart 2: Price Distribution (Histogram)
print("\n")
print("="*60)
print("CHART 2: DISTRIBUTION OF PRODUCT PRICES")
print("="*60)
plt.figure(figsize=(10, 6))
price_data = df_cleaned['Price_Cleaned'].dropna()
plt.hist(price_data, bins=30, color='#e74c3c', edgecolor='black') # Red color
plt.xlabel('Price (INR)', fontsize=12)
plt.ylabel('Number of Products (Frequency)', fontsize=12)
plt.title('Distribution of Mobile Product Prices', fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show() # This displays the chart inline

print("\n--- Analysis and Visualization Complete ---")
```
